
**Apply Edge Detection Operations**

To implement and analyze edge detection techniques using Sobel, Prewitt, and Canny operators on a real image using Python.






```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('/content/Test_1.jpg')

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

**1. Sobel Edge Detection**

```python
sobel_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)

# Magnitude
sobel = cv2.magnitude(sobel_x, sobel_y)
sobel = np.uint8(sobel)
```

**2. Prewitt Edge Detection**

```python
# Define Prewitt kernels
kernelx = np.array([[1, 0, -1],
                    [1, 0, -1],
                    [1, 0, -1]])

kernely = np.array([[1, 1, 1],
                    [0, 0, 0],
                    [-1, -1, -1]])

# Apply filters
prewitt_x = cv2.filter2D(gray, -1, kernelx)
prewitt_y = cv2.filter2D(gray, -1, kernely)

# Magnitude
prewitt = cv2.magnitude(np.float32(prewitt_x), np.float32(prewitt_y))
prewitt = np.uint8(prewitt)
```

**3. Canny Edge Detection**

```python
canny = cv2.Canny(gray, threshold1=100, threshold2=200)
```

```python
plt.figure(figsize=(10, 8))

plt.subplot(2,2,1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(2,2,2)
plt.title('Sobel Edge Detection')
plt.imshow(sobel, cmap='gray')
plt.axis('off')

plt.subplot(2,2,3)
plt.title('Prewitt Edge Detection')
plt.imshow(prewitt, cmap='gray')
plt.axis('off')

plt.subplot(2,2,4)
plt.title('Canny Edge Detection')
plt.imshow(canny, cmap='gray')
plt.axis('off')

plt.show()
```
